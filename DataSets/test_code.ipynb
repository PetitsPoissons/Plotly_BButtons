```python
import os

import pandas as pd
import numpy as np

import sqlalchemy
from sqlalchemy.ext.automap import automap_base
from sqlalchemy.orm import Session
from sqlalchemy import create_engine, inspect
```

```python
#################################################
# Database Setup
#################################################

engine = create_engine("sqlite:///bellybutton.sqlite")
inspector = inspect(engine)
session = Session(engine)

# reflect an existing database into a new model
Base = automap_base()
# reflect the tables
Base.prepare(engine, reflect=True)

#Base.classes.keys()

# Save references to each table
Samples_Metadata = Base.classes.sample_metadata
Samples = Base.classes.samples
```

```python
# see columns in Samples_Metadata
columns_Samples_Metadata = inspector.get_columns('sample_metadata')
for column in columns_Samples_Metadata:
    print(column['name'], column['type'])
```

```python
# see columns in Samples
columns_Samples = inspector.get_columns('samples')
for column in columns_Samples:
    print(column['name'], column['type'])
```

```python
stmt = session.query(Samples).statement
df = pd.read_sql_query(stmt, session.bind)
```
